```python
# Graph:
#     A
#    / \
#   B   C
#  / \  / \
# D   E F  G
#       /
#       H
from collections import deque  # for queue

# Graph
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F','G'],
    'D': [],
    'E': [],
    'F': ['H'],
    'H':[],
    'G':[]
}

def bfs(start):
    visited = set()
    queue = deque([start])   # enqueue start node
    visited.add(start)

    print("BFS Traversal:", end=" ")

    while queue:
        node = queue.popleft()  # dequeue from front
        print(node, end=" ")

        # visit all unvisited neighbors
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)  # enqueue neighbor

# Run BFS starting from A
bfs('A')
```

```python
# Graph:
#     A
#    / \
#   B   C
#  / \  / \
# D   E F  G
#       /
#       H

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F','G'],
    'D': [],
    'E': [],
    'F': ['H'],
    'H':[],
    'G':[]
}

def dfs_with_stack(start):
    visited = set()
    stack = [start]  # push start node

    print("DFS Traversal using Stack:", end=" ")

    while stack:
        node = stack.pop()  # pop top of stack

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            # Push neighbors in reverse to visit left first
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)

# Run DFS starting at A
dfs_with_stack('A')
```
