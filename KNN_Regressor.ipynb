```python
import numpy as np
import pandas as pd

# Example dataset
data = {
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10]
}

df = pd.DataFrame(data)

X = df[["x"]].values   # Features
y = df["y"].values    # Target
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def knn_regression(X_train, y_train, X_test, k=3):
    predictions = []

    for x_test in X_test:
        # Compute distances
        distances = []
        for x_train, y_value in zip(X_train, y_train):
            dist = euclidean_distance(x_test, x_train)
            distances.append((dist, y_value))

        # Sort by distance
        distances.sort(key=lambda x: x[0])

        # Take k nearest neighbors
        k_nearest = distances[:k]

        # Average target values
        prediction = np.mean([value for _, value in k_nearest])
        predictions.append(prediction)

    return np.array(predictions)
```

```python
# New points to predict
X_test = np.array([[1.5], [3.5], [6]])

y_pred = knn_regression(X, y, X_test, k=2)

print("Predictions:", y_pred)
```
